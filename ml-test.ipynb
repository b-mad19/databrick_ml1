Transformers are used to append a new transformed column into the existing column instead of creating a new dataframe.
Binarizer transformer allows us to return a 1 if a value is above a certain threshold or a zero otherwise

TRANSFORMER

```
from pyspark.ml.feature import Binarizer
binarizer=Binarizer(threshold=100, inputCol='Price', outputCol='binarized_price')
display(binarizer.transform(dataframe_name))
```

ESTIMATOR
`StringIndexer` helps usto create an unique value for every unique string that we have.
`OneHotEncoder` estimator helpsto turn that one value into an array

```
from pyspark.ml.feature import StringIndexer
indexer=StringIndexer(inputCol='room_type', outputCol='room_type_index')
```

```
uniqueTypeDF=dataframe_name.select('room_type').distinct() #use distinct value to demonstrate how StringIndexer works
```

**in the `fit` stage we are going to take all the unique values and pair it with some numerical values, and in the `tranform` phase we are appending the column into the dataframe.

```
indexerModel=indexer.fit(uniqueTypeDF)
indexedDF=indexerModel.transform(uniqueTypeDF)
display(indexedDF)
```

OneHotEncoderEstimator: ** here we take that index and turned them into an unique array

```
from myspark.ml.feature import OneHotEncoderEstimator
encoder=OneHotEncoderEstimator(inputCols=['room_type_index'], outputCols=['encoded_room_types'])
encoderModel=encoder.fit(indexedDF)
encodedDF=encoderModel.transform(indexedDF)
display(encodedDF)
```

##PIPELINE

```
from pyspark.ml import Pipeline
pipeline=Pipeline(stages=
                 [binarizer,
                 indexer,
                 encoder]
                 )
```

```
pipelineModel=pipeline.fit(dataframe_name)
transformedDF=pipelineModel.transform(dataframe_name)
display(transformedDF)
```

### `CountVectorizer` : 
Each message or `corpus` is separated into `tokens` and each token occurance is counted and creates a `DTM-Document term Metrics`
   - `document` - single piece of text or tweet or email
   - `corpus`   - collection of `document`, this would be a whole dataset of rows or observations
   - `token`    - each document are split into the individual words known as tokens
   
### `TfidfVectorizer` : 
Instead of filling the DTM with token count its fills up with `Tf-Term frequency` or `idf-inverse document frequency`. `Tfidf` is the product of these two values `Tf` and `idf`
  - `Tf-Term frequency`: weight representing how often a word occurs in a document (text). Higher the term frequency the `Tfidf` increases
  - `idf-Inverse document frequency`: weight representing how common a word is across documents or different texts in the dataframe. If a word is used in many document the **Tfidf** value decreases

```
%python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
   'This is the first document.',
   'This document is the second document.',
   'And this is the third one.',
   'Is this the first document?'
]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names())
print(X.toarray())
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names())
print(X.toarray())
```
